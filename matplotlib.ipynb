Difference between Matplotlib and mplot3d

Matplotlib and mplot3d are both parts of the larger Matplotlib library, but they serve different purposes:

1. Matplotlib: Matplotlib is a comprehensive plotting library for Python that provides a wide range of plotting functionalities. It includes modules for creating 2D plots, such as line plots, scatter plots, bar plots, histograms, etc. Matplotlib is highly customizable and allows you to create publication-quality visualizations with fine-grained control over various aspects of the plot, including axes, labels, colors, legends, and more. It is a versatile library that can be used for both basic and advanced plotting needs.

2. mplot3d: mplot3d is a sub-module of Matplotlib that specifically focuses on 3D plotting capabilities. It provides classes and functions for creating 3D plots, such as surface plots, contour plots, scatter plots, and wireframe plots. The mplot3d module extends Matplotlib's functionality to handle three-dimensional data and visualization. It allows you to create and manipulate 3D axes, apply different projections, set lighting and shading options, and customize the appearance of 3D plots.

We can conclude that Matplotlib is a general-purpose plotting library that covers 2D plotting, while mplot3d is a specialized sub-module within Matplotlib that focuses on 3D plotting. You can use Matplotlib for a wide range of plotting tasks, while mplot3d provides additional functionality and convenience specifically for working with three-dimensional plots.

The `cmap` parameter in Matplotlib's `plot_surface` function specifies the color map to be used for the 3D surface plot. A color map defines the mapping of scalar values to colors.

Here are some commonly used color maps in Matplotlib:

1. `'viridis'`: This is the default color map in Matplotlib. It is a perceptually uniform and visually pleasing color map that ranges from dark blue to yellow.

2. `'plasma'`: This color map also ranges from dark blue to yellow but has a higher contrast than 'viridis'.

3. `'inferno'`: This color map ranges from black to yellow and has a higher contrast than both 'viridis' and 'plasma'.

4. `'magma'`: This color map ranges from black to white and has a high contrast, making it suitable for emphasizing details in the data.

5. `'jet'`: This color map ranges from dark blue to red, with intermediate colors such as cyan and yellow. It was popular in earlier versions of Matplotlib but is no longer recommended due to its perceptual issues.

6. `'cool'`: This color map ranges from cyan to magenta, providing a cool gradient.

7. `'hot'`: This color map ranges from black to red, providing a hot gradient.



```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d

#create coordinate
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

#setup figure and grid 
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

#set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Plot')

plt.show()
```
